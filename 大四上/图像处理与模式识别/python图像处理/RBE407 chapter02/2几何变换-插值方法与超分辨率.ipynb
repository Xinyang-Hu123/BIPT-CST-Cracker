# <center><font color=blue>**图像放大时的插值方法、<font color=red>超分辨率</font>**</font></center>

图像放大时，需要对图像进行**上采样(Up-sampling)**。但问题是新的大图像中会有一些像素在原来的小图像中没有对应的像素,这就需要用插值的方法来确定这些未知像素的值。

图像缩小时，需要对图像进行**下采样(Down-sampling)**。下采样时，需要**抗混叠（Anti-aliasing）**。

<font color=blue>**自学：教材2.1.1 采样，p42－48**</font>

PIL图像缩放函数：
**Image.resize**(size: tuple[int, int], resample: int | None = None)

- size – The requested size in pixels, as a 2-tuple: (width, height).
- resample – An optional resampling filter. This can be one of Resampling.**NEAREST**, Resampling.**BOX**, Resampling.**BILINEAR**, Resampling.**HAMMING**, Resampling.**BICUBIC** or Resampling.**LANCZOS**.

https://pillow.readthedocs.io/en/stable/reference/Image.html#PIL.Image.Image.resize

# <center><font color=blue>**一、将下图用4种插值方法放大3倍，并进行放大效果比较**</font></center>
<br><center>原始图片：clock.jpg，分辨率107x105，宽度107像素，高度105像素，水平分辨率72dpi，垂直分辨率72dpi，位深度24
<br>![原图片](./images/clock.jpg)

## <font color=blue>**PIL 放大图像，插值方法：Nearest Interpolation**</font>

```python
# PIL 放大图像：Nearest Interpolation

from PIL import Image
import matplotlib.pylab as plt

im = Image.open("./images/clock.jpg")
print(im.size, im.width, im.height, im.mode, im.format, type(im))    # 注意：没有im.shape
# (107, 105) 107 105 RGB JPEG <class 'PIL.JpegImagePlugin.JpegImageFile'>
# im.show()

im_large = im.resize((im.width*3, im.height*3), Image.NEAREST)
print(im_large.width, im_large.height)
im_large.save('./images_out/clock_NEAREST.jpg')
# im_large.show()

# 用matplotlib显示
plt.imshow(im,cmap="gray"), plt.axis('off'), plt.title('original'), plt.show()  
plt.imshow(im_large,cmap="gray"), plt.axis('off'), plt.title('zoomed'), plt.show()  
```

## <font color=blue>**PIL 放大图像，插值方法：Bilinear Interpolation**</font>

```python
# PIL 放大图像：Bilinear Interpolation

from PIL import Image
import matplotlib.pylab as plt

im = Image.open("./images/clock.jpg")
print(im.size, im.width, im.height, im.mode, im.format, type(im))    # 注意：没有im.shape
# (107, 105) 107 105 RGB JPEG <class 'PIL.JpegImagePlugin.JpegImageFile'>
# im.show()

im_large = im.resize((im.width*3, im.height*3), Image.BILINEAR) 
print(im_large.width, im_large.height)
im_large.save('./images_out/clock_BILINEAR.jpg')
# im_large.show()

# 用matplotlib显示
plt.imshow(im,cmap="gray"), plt.axis('off'), plt.title('original'), plt.show()  
plt.imshow(im_large,cmap="gray"), plt.axis('off'), plt.title('zoomed'), plt.show()   
```

```python
# [在OpenCV基于深度学习的超分辨率模型实践](https://zhuanlan.zhihu.com/p/671319776)
# 图像放大：EDSR超分辨率插值

import cv2
from cv2 import dnn_superres

print(cv2.__version__, cv2.__file__)
# 4.10.0 /home/dsc/.local/lib/python3.6/site-packages/cv2/__init__.py

# Create an SR object
sr = dnn_superres.DnnSuperResImpl_create()

# Read image
image = cv2.imread('./images/clock.jpg')

# Read the desired model
path = "EDSR_x3.pb"
sr.readModel(path)

# Set the desired model and scale to get correct pre- and post-processing
sr.setModel("edsr", 3)

# Upscale the image
result = sr.upsample(image)


# Save the image
cv2.imwrite("./images_out/clock_upscaled.png", result)

# 在Ubuntu下执行正常
# 4.10.0 /home/dsc/.local/lib/python3.6/site-packages/cv2/__init__.py
# dsc@dsc-Alienware-m17-R3:~$ pip list | grep opencv
# opencv-contrib-python               4.10.0.84
# opencv-python                       4.10.0.84

# 在Windows下执行不正常
# AttributeError: module 'cv2.dnn_superres' has no attribute 'DnnSuperResImpl_create'
# C:\Users\pair_>pip list | find "opencv"
# opencv-contrib-python             4.10.0.84
# opencv-python                     4.8.0.76
# opencv-python-headless            4.10.0.84
# 升级一下OpenCV的版本
# C:\Users\pair_>pip install opencv-python==4.10.0.84
# 估计还是需要重新编译OpenCV（按官方文档进行）！
```

```python
# PIL 放大图像：Lanczos Interpolation

from PIL import Image
# import matplotlib.pylab as plt

im = Image.open("./pics/cracker.png")
print(im.size, im.width, im.height, im.mode, im.format, type(im))    # 注意：没有im.shape
# (107, 105) 107 105 RGB JPEG <class 'PIL.JpegImagePlugin.JpegImageFile'>
# im.show()

im_large = im.resize((im.width*3, im.height*3), Image.NEAREST)
print(im_large.width, im_large.height)
im_large.convert('RGB').save('./pics/cracker_NEAREST.jpg')
# im_large.show()

im_large = im.resize((im.width*3, im.height*3), Image.BILINEAR)
print(im_large.width, im_large.height)
im_large.convert('RGB').save('./pics/cracker_BILINEAR.jpg')
# im_large.show()

im_large = im.resize((im.width*3, im.height*3), Image.BICUBIC)
print(im_large.width, im_large.height)
im_large.convert('RGB').save('./pics/cracker_BICUBIC.jpg')
# im_large.show()

im_large = im.resize((im.width*3, im.height*3), Image.LANCZOS)
print(im_large.width, im_large.height)
im_large.convert('RGB').save('./pics/cracker_LANCZOS.jpg')
# im_large.show()

# # 用matplotlib显示
# plt.imshow(im,cmap="gray"), plt.axis('off'), plt.title('original'), plt.show()  
# plt.imshow(im_large,cmap="gray"), plt.axis('off'), plt.title('zoomed'), plt.show()  
```

```python
# [在OpenCV基于深度学习的超分辨率模型实践](https://zhuanlan.zhihu.com/p/671319776)
# 图像放大：EDSR超分辨率插值

import cv2
from cv2 import dnn_superres

print(cv2.__version__, cv2.__file__)
# 4.10.0 /home/dsc/.local/lib/python3.6/site-packages/cv2/__init__.py

# Create an SR object
sr = dnn_superres.DnnSuperResImpl_create()

# Read image
image = cv2.imread('./pics/cracker.png')

# Read the desired model
path = "EDSR_x3.pb"
sr.readModel(path)

# Set the desired model and scale to get correct pre- and post-processing
sr.setModel("edsr", 3)

# Upscale the image
result = sr.upsample(image)


# Save the image
cv2.imwrite("./pics/cracker_upscaled.png", result)
```
