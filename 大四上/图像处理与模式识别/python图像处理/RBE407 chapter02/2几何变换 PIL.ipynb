## <font color=blue>**Resizing an image  图像缩放**</font>

图像放大时，需要对图像进行**上采样(Up-sampling)**。但问题是新的大图像中会有一些像素在原来的小图像中没有对应的像素,这就需要用插值的方法来确定这些未知像素的值。

图像缩小时，需要对图像进行**下采样(Down-sampling)**。下采样时，需要**抗混叠（Anti-aliasing）**。

<font color=blue>**自学：教材2.1.1 采样，p42－48**</font>

PIL图像缩放函数：
Image.**resize**(size: tuple[int, int], resample: int | None = None)

- size – The requested size in pixels, as a 2-tuple: (width, height).
- resample – An optional resampling filter. This can be one of Resampling.**NEAREST**, Resampling.**BOX**, Resampling.**BILINEAR**, Resampling.**HAMMING**, Resampling.**BICUBIC** or Resampling.**LANCZOS**. 

```python
# 放大图像

im = Image.open("./images/clock.jpg")
print(im.size, im.height, im.width, im.mode, im.format)
# im.show()

im_large = im.resize((im.width*3, im.height*3), Image.BILINEAR) # bi-linear interpolation
print(im_large.width, im_large.height)
# im_large.show()

# 用matplotlib显示
plt.imshow(im,cmap="gray"), plt.title("original clock"), plt.axis('on'), plt.show()  
plt.imshow(im_large,cmap="gray"), plt.title("zoomed clock"), plt.axis('on'), plt.show()  
```

```python
# 缩小图像

im = Image.open("./images/victoria_memorial.png")
print(im.width, im.height)
# im.show()

im_small = im.resize((im.width//5, im.height//5))  #//表示整数除法
print(im_small.width, im_small.height)
# im_small.show()

# 用matplotlib显示
plt.imshow(im,cmap="gray"), plt.title("original victoria memorial"), plt.axis('on'), plt.show()  
plt.imshow(im_small,cmap="gray"), plt.title("small victoria memorial"), plt.axis('on'), plt.show()  
```

## <font color=blue>**Rotating an image  图像旋转**</font>

PIL图像旋转函数：
Image.**rotate**(angle: float, resample: Resampling = Resampling.NEAREST)
<br>Returns a rotated copy of this image. This method returns a copy of this image, rotated the given number of degrees counter clockwise around its centre.

- angle – In degrees counter clockwise.
- resample – An optional resampling filter. This can be one of Resampling.**NEAREST**, Resampling.**BILINEAR**, Resampling.**BICUBIC**. 

```python
im = Image.open("./images/parrot.png") 
print(im.size, im.height, im.width, im.mode, im.format)

im_45 = im.rotate(45) # rotate the image by 45 degrees
# im_45.show() 

# 用matplotlib显示
plt.imshow(im_45), plt.title("rotated parrot"), plt.axis("off")
plt.show()
```

## <font color=blue>**Applying an Affine transformation on an image  图像仿射变换**</font>

Image.**transform**(size: tuple[int, int], method: Transform, data: Sequence[Any] | None = None, resample: int = Resampling.NEAREST) 
<br>Transforms this image. This method creates a new image with the given size, and the same mode as the original, and copies data to the new image using the given transform.

- size – The output size in pixels, as a 2-tuple: (width, height).
- method – The transformation method. This is one of Transform.**EXTENT** (cut out a rectangular subregion), Transform.**AFFINE** (affine transform), Transform.**PERSPECTIVE** (perspective transform), Transform.**QUAD** (map a quadrilateral to a rectangle), or Transform.**MESH** (map a number of source quadrilaterals in one operation).
- data – Extra data to the transformation method.
- resample – Optional resampling filter. It can be one of Resampling.**NEAREST** (use nearest neighbour), Resampling.**BILINEAR** (linear interpolation in a 2x2 environment), or Resampling.**BICUBIC** (cubic spline interpolation in a 4x4 environment).
<br>[源代码](https://pillow.readthedocs.io/en/latest/_modules/PIL/Image.html#Image.transform)

二维仿射变换矩阵T可以应用于图像的每个像素（**在齐次坐标中**），以进行仿射变换，这种变换通常通过反向映射（扭曲）来实现。
A 2-D Affine transformation matrix, T, can be applied on each pixel of an image (**in homogeneous coordinates**) to undergo an Affine transformation, which is often implemented with <font color=blue>**inverse mapping**</font> (warping).
<br>
<br>[How these transformations can be implemented from scratch](https://sandipanweb.wordpress.com/2018/01/21/recursive-graphics-bilinear-interpolation-and-image-transformation-in-python/)
<br>[Resize image with bilinear interpolation in python](https://stackoverflow.com/questions/46631850/resize-image-with-bilinear-interpolation-in-python)
<br>
<br>如下代码所示的是用 shear（剪切）变换矩阵变换输入图像时得到的输出图像。
<br>transform()函数中的数据参数是一个六元组（a，b，c，d，e，f），其中包含仿射变换矩阵（affine transform matrix）的前两行。对于输出图像中的每个像素（x，y），新值
取自输入图像中的位置（ax+by+c，dx+ey+f），使用最接近的像素进行近似。<font color=blue>transform()函数可用于缩放、平移、旋转和剪切原始图像。</font>

```python
im = Image.open("./images/parrot.png") 
print(im.size, im.height, im.width, im.mode, im.format)

out = im.transform((int(1.4*im.width), im.height), Image.AFFINE,data=(1,-0.5,0,0,1,0)) # shear

# 用matplotlib显示
plt.imshow(out), plt.title("shear parrot"), plt.axis("off")
plt.show()
```

## <font color=blue>**Perspective transformation  图像透视变换**</font>

```python
im = Image.open("./images/parrot.png") 
print(im.size, im.height, im.width, im.mode, im.format)

params = [1, 0.1, 0, -0.1, 0.5, 0, -0.005, -0.001]
out = im.transform((im.width//3, im.height), Image.PERSPECTIVE, params,Image.BICUBIC)

# 用matplotlib显示
plt.imshow(out), plt.title("perspective parrot"), plt.axis("off")
plt.show()
```
