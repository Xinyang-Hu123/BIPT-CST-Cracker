```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
import numpy as np
import pandas as pd
```

```python
#导入数据集
mydata = pd.read_csv('C:/Downloads/datasets/iris.csv')
mydata.head()
```

```python
X = mydata.iloc[:,:-1]      #取所有行，取从第0列到最后前一列
T = mydata.iloc[:,-1]       #取所有行，取最后一列

# 将标签转换为数值型（如果尚未转换）
T = pd.Categorical(T).codes

# 转换标签为独热编码
Tx = to_categorical(T)
```

```python
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense

# 定义输入层
input_layer = Input(shape=(4,))

# 添加后续层
x = Dense(4, activation="relu")(input_layer)
output_layer = Dense(3, activation="sigmoid")(x)

# 创建模型
model = Model(inputs=input_layer, outputs=output_layer)

# 余下的编译和训练代码...
```

```python
#查看一下已建好的网络
model.summary()
```

```python
# 编译模型
model.compile(optimizer='adam',
              loss='categorical_crossentropy',  # 对应多分类问题
              metrics=['accuracy'])

# 训练模型
history = model.fit(X, Tx, epochs=1000, validation_split=0.2)  # 可以设置验证集比例

# 查看训练历史中的准确率
accuracy = history.history['accuracy'][-1]
print(f"模型最终训练集准确率为: {accuracy * 100:.2f}%")
```

```python
# 训练模型
history = model.fit(X, Tx, epochs=1000)

# 获取最后一次迭代的准确率
accuracy = history.history['accuracy'][-1]
print(f"模型最终准确率为: {accuracy * 100:.2f}%")

# 如果需要预测和评估模型效果
predictions = model.predict(X)
predicted_classes = np.argmax(predictions, axis=-1)  # 获取预测类别
```
